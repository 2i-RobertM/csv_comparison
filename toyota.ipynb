```python
import pandas as pd
import xlwings as xw

from pathlib import Path
```

```python
#Loading into python dataframe

df_initial = pd.read_csv("toyota_cars.csv")
df_upgrade = pd.read_csv("toyota_cars_upgrade.csv")

#df_initial
#df_upgrade
```

```python
#Dataframes with different shapes so cannot use compare method
print('Shape initial {0} + shape upgrade {1}'.format(df_initial.shape, df_upgrade.shape))
```

```python
#Indexing the information on the updated dataframe to highlight the rows in Excel
df_upgrade = df_upgrade.reset_index()
df_upgrade.head(3)
```

```python
#Merging the two dataframes
df_diff = pd.merge(df_initial, df_upgrade, how = "outer", indicator = "Exist")
#, on ='model' 
print(df_diff)
```

```python
df_diff.shape
```

```python
#Only shows the difference in both dataframes
df_diff_query = df_diff.query("Exist != 'both'")
df_diff_query
```

```python
#Now to highlight the differences
#for this we need to know a row number that includes a difference
#querying the dataframe to only show the updates on our upgrade version 
df_highlight_new = df_diff.query("Exist == 'right_only'")
df_highlight_old = df_diff.query("Exist == 'left_only'")
df_highlight_new
df_highlight_old
```

```python
#Get the row numbers we want to highlight in Excel
highlight_rows_new = df_highlight_new['index'].tolist()
highlight_rows_old = df_highlight_old['index'].tolist()
highlight_rows_old
```

```python
#Converting floats to integers
highlight_rows_new = [int(row) for row in highlight_rows_new]
highlight_rows_old = [int(row) for row in highlight_rows_old]

highlight_rows_old
```

```python
#pandas index starts at 0
# Excel data (w/o header) starts from row 2

first_row_in_excel = 2

highlight_rows_new = [x+ first_row_in_excel for x in highlight_rows_new]
highlight_rows_old = [x+ first_row_in_excel for x in highlight_rows_old]
```

### Using excel wings to highlight the rows

```python
with xw.App(visible = False) as app:
    updated_wb = app.books.open("toyota_cars.csv")
    updated_ws = updated_wb.sheets(1)
    rng = updated_ws.used_range
    
    print(f"USed Range: {rng.address}")
    
    #Hoghlight the rows in excel 
    for row in rng.rows:
        if row.row in highlight_rows_new:
            row.color = (0,255,0) # Green
#         elif row.row in highlight_rows_old:
#             row.color = (255,0,0) # Red
            
            
    updated_wb.save(Path.cwd() / "Difference_Highlighted.xlsx")        
```
